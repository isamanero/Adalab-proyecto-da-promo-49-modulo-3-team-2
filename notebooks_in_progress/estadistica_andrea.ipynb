```python

attrition_counts = df["attrition"].value_counts()

# Crear el gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct="%1.1f%%", startangle=90, colors=["peachpuff", "yellowgreen"])
plt.title("Porcentaje de Continuidad en la Empresa")
plt.show()
```

```python
jobrole_counts = df["jobrole"].value_counts()

# Crear el gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(jobrole_counts, labels=jobrole_counts.index, autopct="%1.1f%%", startangle=90, colors="Set2")
plt.title("Departamentos de la empresa")
plt.show()
```

#### Comparamos jobsatisfaction vs. salary

- cuando las comparamos solo entre si, no se ve una relación clara, todas similares
- en cambio, cuando comparamos salario de las personas que se fueron y las que siguen trabajando, si podemos ver una gran relación, las personas que abandonaron la empresa cobraban menos que las que siguen trabajando



#### NO INCLUIMOS: Comparamos jobsatisfaction vs. distancefromhome 
- no encontramnos relación entre ambas. 

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

sns.barplot(y="distancefromhome", 
            x="jobsatisfaction", 
            data=df, 
            errorbar=('ci', False), 
            hue="attrition", 
            palette="Set2", ax=axes[0])


sns.barplot(y="distancefromhome", 
            x="jobsatisfaction", 
            data=df, 
            errorbar=('ci', False), 
            hue="jobsatisfaction", 
            palette="Set2", ax=axes[1])
;
```

#### Comparamos jobsatisfaction vs. joblevel

- No es relevante el nivel del puesto de trabajo con la satisfacción cuando comparamos globalmemte
- en cambio, si vemos que tiene una gran relevancia cuando introducimos la variable "attrition"

Podemos deducir que para las personas que ya no trabajan en la empresa, el nivel de satisfacción era menor cuanto menor era el nivel de su puesto de trabajo


```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

sns.barplot(x="jobsatisfaction",
            y="joblevel", data=df,
            ax=axes[0], palette="Set2", hue="jobsatisfaction")

sns.barplot(x="jobsatisfaction",
            y="joblevel", data=df,
            ax=axes[1], palette="Set2",
            hue="attrition")

#plt.subplots_adjust(wspace=0.5)
```

#### Comparamos jobsatisfaction vs. department

- Por departamentos si se ve relacion, HHRR está menos satisfecho, y esto queda aun más visible cuando introducimos la variable "attrition".

Podemos deducir que el departamento menos satisfecho es HHRR, sobre todo para las personas que ya no trabajan en la empresa, quizá se han realizado algunos cambios posteriores a la salida de estas personas que han mejorado la valoración global de satisfacción. Analizar más... RRHH tienen la mayor diferencia de satisfacción entre quienes se quedan y se van. 

Hasta el momento, podemos ver que la variable de satisfacción del puesto de trabajo, es importante, que la empresa podría renovar constantemente estas encuestas de satisfacción del puesto de trabajo.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

sns.barplot(x="jobsatisfaction",
            y="department", data=df,
            ax=axes[0], palette="Set2")

sns.barplot(x="jobsatisfaction",
            y="department", data=df,
            ax=axes[1], palette="Set2",
            hue="attrition")
```

### NO LA INCLUIMOS: comparamos jobsatisfaction vs. datebirth 
no vemos relación, faltan datos por edades

```python
# Convertir a formato de fecha
df["datebirth"] = pd.to_datetime(df["datebirth"]) 
df["age"] = pd.to_datetime("today").year - df["datebirth"].dt.year 
```

```python
print(df[["age", "jobsatisfaction"]].corr())
```

```python
df["age_group"] = pd.cut(df["age"], bins=[18, 30, 40, 50, 60, 70], labels=["18-30", "31-40", "41-50", "51-60", "61-70"])

sns.boxplot(y="age_group", x="jobsatisfaction", data=df)
plt.title("Satisfacción laboral por grupos de edad")
plt.show()
```
