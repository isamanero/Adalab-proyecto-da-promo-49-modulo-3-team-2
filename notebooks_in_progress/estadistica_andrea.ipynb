#### Comparamos jobsatisfaction vs. salary

- cuando las comparamos solo entre si, no se ve una relación clara, todas similares
- en cambio, cuando comparamos salario de las personas que se fueron y las que siguen trabajando, si podemos ver una gran relación, las personas que abandonaron la empresa cobraban menos que las que siguen trabajando



```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

sns.barplot(y="salary", 
            x="jobsatisfaction", 
            data=df, 
            errorbar=('ci', False), 
            hue="jobsatisfaction", 
            palette="Set2", 
            ax=axes[1])

sns.barplot(y="salary", 
            x="jobsatisfaction", 
            data=df, 
            errorbar=('ci', False), 
            hue="attrition", 
            palette="Set2", 
            ax=axes[0])
```

#### Comparamos jobsatisfaction vs. joblevel

- No es relevante el nivel del puesto de trabajo con la satisfacción cuando comparamos globalmemte
- en cambio, si vemos que tiene una gran relevancia cuando introducimos la variable "attrition"

Podemos deducir que para las personas que ya no trabajan en la empresa, el nivel de satisfacción era menor cuanto menor era el nivel de su puesto de trabajo


```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

sns.barplot(x="jobsatisfaction",
            y="joblevel", data=df,
            ax=axes[0], palette="Set2", hue="jobsatisfaction")

sns.barplot(x="jobsatisfaction",
            y="joblevel", data=df,
            ax=axes[1], palette="Set2",
            hue="attrition")

#plt.subplots_adjust(wspace=0.5)
```

### comparamos jobsatisfaction vs. datebirth 
no vemos relación, faltan datos por edades

```python
# Convertir a formato de fecha
df["datebirth"] = pd.to_datetime(df["datebirth"]) 
df["age"] = pd.to_datetime("today").year - df["datebirth"].dt.year 
```

```python
print(df[["age", "jobsatisfaction"]].corr())
```

```python
df["age_group"] = pd.cut(df["age"], bins=[18, 30, 40, 50, 60, 70], labels=["18-30", "31-40", "41-50", "51-60", "61-70"])

sns.boxplot(y="age_group", x="jobsatisfaction", data=df)
plt.title("Satisfacción laboral por grupos de edad")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

sns.barplot(x="jobsatisfaction",
            y="datebirth", data=df,
            ax=axes[0], palette="Set2")

sns.barplot(x="jobsatisfaction",
            y="datebirth", data=df,
            ax=axes[1], palette="Set2",
            hue="attrition")
```
