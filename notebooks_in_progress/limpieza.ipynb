#### pasar a float

OK, OK, OK

```python
# para: Monthlyincome, Monthlyrate, Salary
# vamos a hacer una función para pasar estos datos que son str a float y quitar el simbolo $:

def pasar_a_float(dato):
                     
    try:
        dato = dato.replace(',', '.').replace("$", "")
        return float(dato)
    except:
        return np.nan
```

```python
lista_dolares = ["monthlyincome", "monthlyrate", "salary"]
```

```python
for c in lista_dolares:
    df1[c] = df1[c].apply(pasar_a_float)
```
