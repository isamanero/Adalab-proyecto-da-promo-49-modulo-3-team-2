### DAILYRATE
Redondear a 2 decimales

```python
df["dailyrate"] = df["dailyrate"].round(2)
df["dailyrate"].head(3)
```

### DISTANCEFROMHOME
Eliminar el "-" porque tenemos valores negativos

```python

df["distancefromhome"] = df["distancefromhome"].map(abs) # Nos saca el valor absoluto
df["distancefromhome"].sample(10)
```

### GENDER

Valores 0 y 1, remplazarlos por Male y Female. (0 Male, 1 Female)

```python
# Modifica la columna de INT a STR
df["gender"]=df["gender"].astype(str)

#Sustituimos por los valores correctos
df['gender'] = df['gender'].replace({'0': 'Male', '1': 'Female'})

print(f"Todos los valores han sido sustituidos por {df['gender'].unique()}\n-----")
```

### JOBROLE

Cambiar todas las palabras con la primera en mayúscula y eliminar los espacios a los lados

```python
df["jobrole"]=df["jobrole"].str.title().str.strip()
df["jobrole"].unique()
```

### PERFORMANCERATING, TOTALWORKINGHOURS, WORKLIFEBALANCE, SALARY, MONTHLYINCOME, MONTHLYRATE

- Sustituir la ',' por el '.'
- Eliminar el símbolo del $
- Convertir de STR a FLOAT

```python
def pasar_a_float(dato):
    """ Función para convertir STR a FLOAT y sustituir la ',' por el '.'"""                 
    try:
        dato = dato.replace(',', '.').replace("$", "")
        return float(dato)
    except:
        return np.nan
```

```python
columnas_numericas = ["performancerating", 
                      "totalworkingyears", 
                      "worklifebalance", 
                      "salary", 
                      "monthlyincome", 
                      "monthlyrate"]

for c in columnas_numericas:
    df[c] = df[c].apply(pasar_a_float)
```

```python
# Comprobamos que los datos son correctos después del cambio
df[columnas_numericas].sample(3)
```

```python
df[columnas_numericas].info()
```

### REMOTEWORK

Sustituir 1/0 y True/False por ->  Yes/No


```python
print(f"Los valores que contiene remotework son : {df["remotework"].unique()}\n-----")

remote_map = { "Yes": "Yes", "1": "Yes", "True": "Yes", "0": "No", "False": "No"} 
df["remotework"] = df["remotework"].map(remote_map)

print(f"Los valores que contiene remotework una vez sustituidos son : {df["remotework"].unique()}\n-----")
```
