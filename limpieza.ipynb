```python
# Comenzamos por la columna gender:
# Valores 0 y 1, remplazarlos por Male y Female. (0 Male, 1 Female)

#usamos df1 que es nuetro data con columnas borradas

df1["gender"]=df1["gender"].astype(str) #modifica la columna de int a str
```

```python
df1["gender"].info()
```

```python
df1['gender'] = df1['gender'].replace({'0': 'Male', '1': 'Female'})
```

```python
df1
```

```python
# Dailyrate: Redondear a 2 decimales

df1["dailyrate"]=df1["dailyrate"].round(2)
```

```python
# Distancefromhome: vamos a quitar el guión porque tenemos valores negativos( es un int)

df1["distancefromhome"]=df1["distancefromhome"].map(abs) # Nos saca el valor absoluto
df1
```

```python
# jobrole: Mayúsculas y minúsculas intercaladas. Poner las palabras con la primera en mayúscula

df1["jobrole"]=df1["jobrole"].str.title()
```

```python
# Creamos funcion para pasar a int para columnas "performancerating","totalworkingyears","worklifebalance": 

def pasar_a_int(dato):
                     
    try:
        dato = dato.replace(',', '.')
        dato2 = float(dato)
        return int(dato2)
    except:
        return np.nan
```

```python
columnas_a_int = ["performancerating","totalworkingyears","worklifebalance"]
```

```python
for c in columnas_a_int:
    
        df1[c] = df1[c].apply(pasar_a_int) #Con esta función pasamos nuestras 3 columnas a float
    
```
